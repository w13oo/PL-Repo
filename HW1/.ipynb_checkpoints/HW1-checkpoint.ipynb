```python
import pandas as pd
```

```python
data = pd.read_csv('Number_of_visitors_in_Tainan.csv', header=None)
data.head()
```

```python
data.columns = ['觀光遊憩區別', '遊客人次有門票_需購票', '遊客人次無門票_免費', '遊客人次假日', '遊客人次非假日',
              '門票收入_元', '上年同月遊客人數', '備註', ]
```

```python
data.head()
```

```python
print(data['遊客人次有門票_需購票'])
```

```python
1
#將 遊客人次非假日 排名
data_sorted = data.sort_values('遊客人次非假日')
data_sorted
```

```python
2
#將 遊客人次假日 排名
data_sorted = data.sort_values('遊客人次假日')
data_sorted
```

```python
3
#將 上年同月遊客人數 以升序排列 (由小到大)
data_sorted = data.sort_values('上年同月遊客人數')
data_sorted
```

```python
4
#將 遊客人次無門票_免費 以降序排列 (由大到小)
sorted_data = data.sort_values(by='遊客人次無門票_免費', ascending=False)
print(sorted_data[['觀光遊憩區別', '遊客人次無門票_免費']])
```

```python
5
max_index = data['上年同月遊客人數'].idxmax()
max_value = data.loc[max_index, '上年同月遊客人數']
max_area = data.loc[max_index, '觀光遊憩區別']

print('上年同月遊客人數的最大值為：', max_value)
print('最大值對應的觀光遊憩區為：', max_area)


```

```python

# 按照觀光遊憩區別進行分組，計算平均值
grouped_data = data.groupby('觀光遊憩區別').mean()

# 輸出結果
print(grouped_data)
```

```python
6
# 按照觀光遊憩區別進行分組，比較假日和非假日遊客人次的大小
grouped_data = data.groupby('觀光遊憩區別').apply(lambda x: '假日遊客較多' if x['遊客人次假日'].sum() > x['遊客人次非假日'].sum() else '非假日遊客較多')

print(grouped_data)
```

```python
7
# 將所有遊客人次假日相加
假日遊客總數 = data['遊客人次假日'].sum()

# 輸出結果
print('假日遊客總數為：', 假日遊客總數)
```

```python
8
# 將所有遊客人次非假日相加
非假日遊客總數 = data['遊客人次非假日'].sum()

# 輸出結果
print('非假日遊客總數為：', 非假日遊客總數)
```

```python
9
#計算遊客人次總合並將其升序排列
grouped_data = data.groupby('觀光遊憩區別').sum()
grouped_data['Total'] = grouped_data['遊客人次假日'] + grouped_data['遊客人次非假日']
sorted_data = grouped_data.sort_values('Total', ascending=False)
print(sorted_data)
```

```python
10
#列出遊客總數前五名
top_five = sorted_data.head(5)
print('遊客總數前五名:')
for index, row in top_five.iterrows():
    print(f'{index}: {row["遊客人次假日"]} (遊客人次假日) + {row["遊客人次非假日"]} (遊客人次非假日) = {row["Total"]}')
```
