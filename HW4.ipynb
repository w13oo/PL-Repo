```python
import re
import requests

url = "https://www.gq.com.tw/entertainment/article/%E9%98%BF%E5%87%A1%E9%81%94-%E6%B0%B4%E4%B9%8B%E9%81%93-%E7%9C%8B%E9%BB%9E-%E7%BC%BA%E9%BB%9E"
response = requests.get(url)
text = response.text

# 使用正則表達式找出所有中文字符
pattern = "[\u4e00-\u9fa5]+"
chinese_text = re.findall(pattern, text)

# 輸出結果
print(chinese_text)
```

```python
import jieba
import jieba.analyse
import pandas as pd
import numpy as np
from collections import Counter
```

```python
# 測試結巴功能
documents = ["我的名字是王于萱"]
# 精確模式
for sentence in documents:
    seg_list = jieba.cut(sentence)
    print('/'.join(seg_list))
```

```python
allwords = []
for content in chinese_text:
    allwords.append(content)
allwords
```

```python
alljie = [] 
for i in allwords:
    seg = jieba.analyse.extract_tags(i, 50)
    alljie.append(seg)
alljie
```

```python
import numpy
from PIL import Image
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
flat = [x for sublist in alljie for x in sublist]
diction = Counter(flat)
font = "Arial Unicode"
mask = numpy.array(Image.open("20200511-1-600x300.jpg"))
wordcloud = WordCloud(background_color = "white", font_path=font, mask = mask)#, mask = mask)
wordcloud.generate_from_frequencies(diction)
plt.figure(figsize=(400,400))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
